```python
# importamos la librería que usaremos para leer y limpiar los datos
import pandas as pd
```

```python
# la funcion "read_csv" lee la info del database y lo guarda en una variable
data_base = pd.read_csv("/content/Basura-oceanos.csv")
data_base
```

```python
# la columna "Entity" pasa a ser "Productos"
data_base.rename(columns={"Entity": "Productos"}, inplace=True)
data_base
# la columna "East Asia and Pacific" pasa a ser "Asia Oriental y el Pacífico"
data_base.rename(columns={"East Asia and Pacific": "Asia Oriental y el Pacífico"}, inplace=True)
data_base
# la columna "South Asia" pasa a ser "Asia del sur"
data_base.rename(columns={"South Asia": "Asia del sur"}, inplace=True)
data_base
# la columna "Latin America and the Carribean" pasa a ser "Latinoamerica y el Caribe"
data_base.rename(columns={"Latin America and the Carribean": "Latinoamerica y el Caribe"}, inplace=True)
data_base
# la columna "Sub-Saharan Africa" pasa a ser "Africa Subsahariana"
data_base.rename(columns={"Sub-Saharan Africa": "Africa Subsahariana"}, inplace=True)
data_base
# la columna "North Africa and Middle East" pasa a ser "Africa del norte y Medio Oriente"
data_base.rename(columns={"North Africa and Middle East": "Africa del norte y Medio Oriente"}, inplace=True)
data_base
```

```python
# la función "isnull().sum()" entrega el nº de datos nulos en cada columna
data_base.isnull().sum()
```

```python
# la función "unique()" entrega el listado de valores únicos de la columna
data_base["Code"].unique()
```

```python
# La función "drop" con "axis=1" elimina la columna que se le entrega
data_base=data_base.drop("Code", axis=1)
data_base
```

```python
# La función "drop" con "axis=1" elimina la columna que se le entrega
data_base=data_base.drop("High income", axis=1)
data_base
```

```python
# La función "drop" con "axis=1" elimina la columna que se le entrega
data_base=data_base.drop("Year", axis=1)
data_base
```

```python
data_base = data_base.drop([1, 2, 3], axis=0)
data_base
```

```python
data_base = data_base.drop([4, 5, 6, 13, 15], axis=0)
data_base
```

```python
# Remplazamos los valores nulos por cero ya que significa que hay 0% de emisión de basura de ese producto
data_base = data_base.fillna(0)
data_base
```
